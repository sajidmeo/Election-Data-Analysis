```python
total = []
male = []
fmale = []
young_fmale = []
dataf = pd.DataFrame()
for i in data['poll_booth'].unique():
    pool = data.loc[data['poll_booth'] == i,:]
    female = pool.loc[pool['Gender']=='F',:]
    young_female = female.loc[female['Age']<=35,:]
    
    print("poll_booth no. = ",i,"total people = ",len(pool),"Male =",len(pool.loc[pool['Gender']=='M',:]), "FeMale =",len(pool.loc[pool['Gender']=='F',:]),"Average Age =",pool['Age'].mean(),"meadian age = ",pool['Age'].median(),"percentage of younger female = ",len(young_female)/len(female)*100)
    #total_people['poll] = i
    total.append(i)
    male.append(len(pool.loc[pool['Gender']=='M',:]))
    fmale.append(len(pool.loc[pool['Gender']=='F',:]))
    young_fmale.append(len(young_female)/len(female)*100)
    #total_people[i] = len(pool)
dataf['part_no.'] = total
dataf['male'] = male
dataf['femal'] = fmale
dataf['young_female']= young_fmale
dataf
dataf.to_csv('young_female.csv')
```

```python
def prep_for_bar(x,data):
    
    a =[]
    for i in x:
        print("vote percentage of ",i,"is :",data[i].iloc[174]/113729*100)
        a.append(data[i].iloc[174]/113729*100)
    freq_series = pd.Series.from_array(a)
    return freq_series
```

```python
freq_series = prep_for_bar(candidate_2015,data_2015)
plt.figure(figsize=(12, 8))
ax = freq_series.plot(kind='bar')
ax.set_title('percentage of votes (2015)')
#ax.set_xlabel('Amount ($)')
#ax.set_ylabel('Frequency')
ax.set_xticklabels(candidate_2015)
```
